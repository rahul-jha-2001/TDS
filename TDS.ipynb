```python
def nested_data(dic:dict,li:list):
    data = {}
    for i in li:
        if i == "full_name":
            data[i] = dic.get("full_name")
        if i == "created_at":
            data[i] = dic.get("created_at")
        if i == "stargazers_count":
            data[i] = dic.get("stargazers_count")
        if i == "watchers_count":
            data[i] = dic.get("watchers_count")
        if i == "language":
            data[i] = dic.get("language")
        if i == "has_projects":
            data[i] = dic.get("has_projects")
        if i == "has_wiki":
            data[i] = dic.get("has_wiki")
        if i == "license":
            if dic.get("license") != None:
                data["license_name"] = dic["license"]["key"]
                continue
            data["license_name"] = None   
    return data

required_fields  = ["full_name","created_at","stargazers_count","watchers_count","language",
                "has_projects","has_wiki","license"]

repos = []

Directory = os.path.join(os.getcwd(),"Repos")

for i in os.listdir(Directory):
    file_path = os.path.join(Directory,i)
    with open(file_path,"r") as f:
        js = json.load(f)
        for j in js:
            data = nested_data(j,required_fields)
            repos.append(data)
df = pd.DataFrame(repos)
df.to_csv("repositories.csv")
```
